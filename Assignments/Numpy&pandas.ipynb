```python
data2 = pd.DataFrame({'Brand' : ['Maruti', 'Hyundai', 'Tata', 
                                'Mahindra', 'Maruti', 'Hyundai', 
                                'Renault', 'Tata'], 
                     'Year' : [2012, 2014, 2011, 2015, 2012,  
                               2016, 2014, 2018], 
                     'Kms Driven' : [50000, 30000, 60000,  
                                     25000, 10000, 46000,  
                                     31000, 15000], 
                     'City' : ['Gurgaon', 'Delhi', 'Mumbai',  
                               'Delhi', 'Mumbai', 'Delhi',  
                               'Mumbai','Chennai'], 
                     'Mileage' :  [28, 27, 25, 26, 28, 29, 24, 21]}) 

data2
```

```python
display(data2.loc[(data2.Brand == 'Maruti') & (data2.Mileage > 25)])
```

```python
display(data2.loc[2 : 5])
```

```python
# updating values of Mileage if Year < 2016 
data2.loc[(data2.Year < 2016), ['Mileage']] = 22
display(data2)
```
